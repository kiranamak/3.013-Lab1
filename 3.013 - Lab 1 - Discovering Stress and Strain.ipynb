```python
import matplotlib.pyplot as plt
import numpy as np
```

## Second Stretch

```python
initial = np.array([31,71,101,121])
observed = np.array([22,52,75,90])
```

```python
m, b = np.polyfit(initial,observed,1)
```

```python
plt.scatter(initial,90/120*initial, label='theory')
plt.scatter(initial,observed, label='observed')
plt.plot(initial, m*initial + b, label='best fit line')
plt.legend()
plt.xlabel('Initial distance from line B (mm)')
plt.ylabel('Displacement (mm)')
plt.title('Thin Stretch: Displacement of Stretched Rubber vs. Initial position')
plt.show()
```

```python
print(f'slope: {m}')
print(f'y-intercept: {b}')
```

The equation for the line of best fit is $Displacement = 0.757 \cdot Position_{initial} - 1.52$. The expected slope is $90/121=0.744$. This corresponds to an error of $1.75\%$.

## Third Stretch

```python
initial_thick = np.array([30, 70, 109])
initial_thin = initial

observed_thick = np.array([5,12,16])
observed_thin = np.array([17, 37, 53, 63])
```

```python
m_thick, b_thick = np.polyfit(initial_thick,observed_thick,1)
m_thin, b_thin = np.polyfit(initial_thin,observed_thin,1)
```

```python
plt.scatter(initial_thick, observed_thick, label = 'observed (thick)')
plt.scatter(initial_thin, observed_thin, label = 'observed (thin)')

plt.plot(initial_thick, m_thick*initial_thick + b_thick, label = 'best fit line (thick)')
plt.plot(initial_thin, m_thin*initial_thin + b_thin, label = 'best fit line (thin)')

plt.plot(initial_thick, 0.25 * initial_thick, label = 'theory (thick)')
plt.plot(initial_thin, 0.5 * initial_thin, label = 'theory (thin)')


plt.xlabel('Initial distance from start of thick/thin section (mm)')
plt.ylabel('Displacement (mm)')
plt.title('Full Stretch: Displacement of Stretched Rubber vs. Initial position')

plt.legend()
plt.show()
```

```python
print(f'Line of best fit for thick section: Displacement = {m_thick}*d + {b_thick}')
print(f'Line of best fit for thin section: Displacement = {m_thin}*d + {b_thin}')
```

We expect the slope of the thick section to be 0.25 so this value is significantly off, likely due to the age of the rubber. The think section is has about a 2.6% error (it is expected to be 0.5).

## Final Stretch

```python
force = np.array([2,4,6,8,10])
final_length = np.array([44,46,49,54,60])

strain = (final_length - 40)/40
stress = force/(0.04*0.015)
```

```python
m,b = np.polyfit(strain,stress,1)
```

```python
plt.scatter(strain, stress, label='observed')
plt.plot(strain, m*strain + b, label='best fit line')

plt.xlabel('Strain')
plt.ylabel('Stress (N/m^2)')
plt.title('Stress vs. Strain Curve for Rubber Sample')
plt.legend()

plt.show()
```

```python
print(f'Line of best fit: Stress = {m}*strain + {b}')
```

The slope is $\approx 3.2 \times 10^4 N \cdot m^{-2}$  which converts to $0.032 MPa$.
